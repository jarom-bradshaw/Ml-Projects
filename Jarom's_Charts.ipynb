```
# Data Exploration
# Graphs
import seaborn as sns
import matplotlib.pyplot as plt
```

```
# Helpful Resources:

# Countplots versus barplots:
# ci = confidence interval
# kind = kind of chart
# data = the dataframe you are equaling it to
# x= x-axis
# y=y-axis

# https://campus.datacamp.com/courses/introduction-to-data-visualization-with-seaborn/visualizing-a-categorical-and-a-quantitative-variable?ex=1#:~:text=Bar%20plots%20look%20similar%20to,each%20day%20of%20the%20week.


# # Data Exploration
# # 1. Previous Campaign Outcome (poutcome) vs. Subscription Outcome (y)
# Type: Countplot or Barplot
# Purpose: To assess whether previous campaign outcomes (success, failure,
# nonexistent) have a direct correlation with the likelihood of subscribing to a
# term deposit in the current campaign. This will be important for identifying
# customers who have responded well to past marketing efforts and may be more
# likely to subscribe again.

sns.catplot(x="poutcome", hue="y", data=df, kind="count")
plt.title("Previous Campaign Outcome vs. Subscription Outcome")

# 2. Number of Contacts in Previous Campaign (previous) vs. Subscription Outcome
# (y) Type: Boxplot or Violin plot
# Purpose: To analyze whether the number of previous contacts affects the
# current campaign's outcome. Customers who were contacted more frequently in
# previous campaigns might have a lower likelihood of subscribing, while those
# with fewer prior contacts might be more receptive.
sns.boxplot(x="y", y="previous", data = df)
plt.title("Number of Contacts in Previous Campaign vs. Subscription Outcome")

# 3. Contact Type (contact) vs. Subscription Outcome (y)
# Type: Countplot or Barplot
# Purpose: To see if certain contact methods (cellular vs. telephone) have
# different success rates for gaining subscriptions, which could guide the bank
# on the most effective communication channels.
sns.catplot(x="contact", hue="y", data=df, kind="count")
plt.title("Type of Last Contact vs. Subscription Outcome")

# 4. Month of Last Contact (month) vs. Subscription Outcome (y)
# Type: Countplot
# Purpose: To explore whether certain months (e.g., holiday months or specific
# seasons) show better results in terms of subscription. This could reveal
# seasonal trends in customer behavior.
sns.catplot(x="month", hue="y", data=df, kind="count")
plt.title("Month of Last Contact vs. Subscription Outcome")

#
# I think the data does not have jan feb, mar?
#

# 5. Days of the Week of Last Contact (dayofweek) vs. Subscription Outcome (y)
# Type: Countplot or Barplot
# Purpose: To investigate if there are specific days of the week that lead to
# better results for the campaign. This could give insights into the best days
# for contacting customers based on historical patterns.
sns.catplot(x="day_of_week", hue="y", data=df, kind="count")
plt.title("Days of the Week of Last Contact vs. Subscription Outcome")

# 6. Economic Conditions and Subscription Outcome (emp.var.rate, cons.conf.idx,
# etc.)
# Type: Pairplot or Heatmap (Correlation Matrix)
# Purpose: To assess how various economic indicators (employment variation rate,
# consumer confidence, etc.) correlate with subscription outcomes. This could
# provide insights into whether external economic conditions influence customer
# decision-making.
# Displayed in other code snippet

# economic_factors = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
# corr_matrix = df[['economic_factors', 'y']]

# corr_matrix = df[['emp.var.rate', 'cons.conf.idx', 'cons.price.idx', 'euribor3m', 'nr.employed', 'y']]
# corr_matrix['y'] = df['y'].map({'yes': 1, 'no':0})

# corr_matrix = corr_matrix.corr()

# sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
# plt.title("Correlation between Economic Factors and Subscription Outcome")
```

```
# Heatmap but easier with dictionary.
import numpy as np

# Labels for economic factors and subscription outcome (for this question)
economic_labels_1 = {
    'emp.var.rate': 'Employment Variation Rate',
    'cons.price.idx': 'Consumer Price Index',
    'cons.conf.idx': 'Consumer Confidence Index',
    'euribor3m': 'Euribor 3-Month Rate',
    'nr.employed': 'Number of Employees',
    'y': 'Subscription Outcome'
}

# Create the correlation matrix with the economic factors and subscription outcome
corr_matrix1 = df[['emp.var.rate', 'cons.conf.idx', 'cons.price.idx', 'euribor3m', 'nr.employed', 'y']]

# Map 'y' from 'yes'/'no' to 1/0 for correlation calculation
corr_matrix1['y'] = df['y'].map({'yes': 1, 'no': 0})

# Compute the correlation matrix
corr_matrix1 = corr_matrix1.corr()

# Rename columns and indices for easier readability in the heatmap
corr_matrix1.rename(columns=economic_labels_1, inplace=True)
corr_matrix1.rename(index=economic_labels_1, inplace=True)

# Create the mask to hide the upper triangle
mask = np.triu(np.ones_like(corr_matrix1, dtype=bool))

# Plot the heatmap
# sns.heatmap(corr_matrix1, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={'label': 'Correlation Coefficient'})
# Plot the heatmap with the mask
sns.heatmap(
    corr_matrix1,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    cbar_kws={'label': 'Correlation Coefficient'},
    mask=mask
)
plt.title("Correlation between Economic Factors and Subscription Outcome")
plt.show()
```

```
# 7. Number of Contacts (campaign + previous) vs. Subscription Outcome (y)
# Type: Scatterplot or Boxplot
# Purpose: To examine the combined total number of contacts (previous + current
# campaign) and see how it affects subscription rates. This can help determine
# if more contacts improve the chances of a successful outcome or if
# over-contacting leads to diminishing returns.
df['total_contacts'] = df['campaign'] + df['previous']
sns.boxplot(x="y", y="total_contacts", data=df)
plt.title("Total Number of Contacts vs. Subscription Outcome")

# 8. Customer Segments Based on Job and Subscription Outcome
# Type: Countplot or Barplot
# Purpose: To identify which job categories (e.g., admin, technician, student,
# etc.) are more likely to subscribe. This could help in creating customer
# segments for targeted marketing campaigns.
sns.catplot(x="job", hue="y", data=df, kind="count")
plt.xticks(rotation=90)
plt.title("Customer Segments Based on Job vs. Subscription Outcome")

# 9. Correlation between Economic Factors
# Type: Heatmap
# Purpose: To explore relationships between key economic factors
# (e.g., emp.var.rate, cons.price.idx, euribor3m) and identify any strong
# correlations. Understanding these relationships could help the bank understand
# if certain economic trends impact customer behavior and subscription
# likelihood.
# economic_factors = ['emp.var.rate', 'cons.price.idx', 'cons.conf.idx', 'euribor3m', 'nr.employed']
# corr_matrix = df[economic_factors].corr()
# sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={'label': 'Correlation Coefficient'})
# plt.title("Correlation between Economic Factors")
# Displayed in other code snippet
```
